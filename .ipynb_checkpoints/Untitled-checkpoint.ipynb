```python
import tensorflow as tf
import numpy as np 
import matplotlib.pyplot as plt 

class Training:
    def __init__(self):
        self.datasetPath = 'D:\\Datasets\\Apple'
        self.imgShape = (64,64)
        
    def getData(self,subset='training'):
        try:
            trainingData = tf.keras.utils.image_dataset_from_directory(
                self.datasetPath,
                validation_split=0.2,
                subset=subset,
                image_size= self.imgShape,
                seed = 58,
                batch_size = 32
            )
            return trainingData
        except Exception as e:
            raise e
    
    def visualizeData(self):
        try:
            trainingData = self.getData()
            classNames = trainingData.class_names
            for images,lables in trainingData.take(1):
                plt.figure(figsize=(10,10))
                for i in range(18):
                    plt.subplot(6,3,i+1)
                    plt.imshow(images[i].numpy().astype('uint32'))
                    plt.title(classNames[lables[i].numpy().astype('uint8')])
        except Exception as e:
            raise e
                  
    def createModel(self):
        try:
            AUTOTUNE = tf.data.AUTOTUNE
            trainingData = self.getData();
            valData = self.getData(subset='validation')
            CLASSES = len(trainingData.class_names)
            trainingData = trainingData.cache().prefetch(buffer_size=AUTOTUNE)
            valData = valData.cache().prefetch(buffer_size=AUTOTUNE)
            
            
            LAYERS = [
                tf.keras.layers.Rescaling(1./255),
                tf.keras.layers.Conv2D(32,3,activation='relu'),
                tf.keras.layers.MaxPooling2D(),
                tf.keras.layers.Conv2D(32,3,activation='relu'),
                tf.keras.layers.MaxPooling2D(),
                tf.keras.layers.Conv2D(32,3,activation='relu'),
                tf.keras.layers.MaxPooling2D(),
                tf.keras.layers.Flatten(),
#                 tf.keras.layers.Dense(256,activation='relu'),
                tf.keras.layers.Dense(128,activation='relu'),
                tf.keras.layers.Dense(CLASSES)
            ]
            
            model = tf.keras.models.Sequential(LAYERS)
            model.compile(
                optimizer='adam',
                metrics=['accuracy'],
                loss='sparse_categorical_crossentropy'
            )
            
            hist=model.fit(
                trainingData,
                epochs=100,
                validation_data=valData
            )
            model.save("fruitLife.h5")
        except Exception as e:
            raise e
            
            
              
obj = Training()
obj.createModel()
    
```
